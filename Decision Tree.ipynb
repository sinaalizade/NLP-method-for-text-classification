### My First ML Model

#### Showing Columns of CSV file

```python
mh.columns
```

```python
y = mh.Price
y
```

```python
mh.mean(axis = 1)
```

```python
mh_features = ['Rooms', 'Bathroom',
                      'Landsize', 'Lattitude', 'Longtitude']
```

```python
x = mh[mh_features]
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
ml_model= DecisionTreeRegressor(random_state=1)
ml_model.fit(x,y)
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(ml_model.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = ml_model.predict(x)
mean_absolute_error(y,predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x,train_y,val_y = train_test_split(x,y,random_state =0)
ml_model = DecisionTreeRegressor()
ml_model.fit(train_x,train_y)
val_predictions = ml_model.predict(val_x)
print(mean_absolute_error(val_y, val_predictions))
```
